<a id='4'></a>
**Question 4:** 

What is the percent of total Sales for the ‘Home Office’ Customer Segment in July of 2012?

**Solution Steps:** 

* Filter “Order Date” to 2012
* 2nd filter "Order Date" to July
* Drag “Customer Segment” to Columns 
* Drag "Sales" to Rows (right click, quick table calculations, percent of total)
* Click check box for show mark labels on Label 

**Answer:** 

b. 23.97%

<a id='5'></a>
**Question 5:** 

Find the top 10 Product Names by Sales within each region. Which product is ranked # 2 in both the Central & West regions in 2011? 

**Solution Steps:**

* Filter “Product Name”, Top tab, By field, Top 10 Sum of Sales, click OK
* 2nd filter “Order Date” to 2011 
* 3rd filter “Region” to Central and West
* Drag “Product Name” to Columns
* Drag “Sales” to Rows, click, quick table calculation, select Rank  

**Answer**: 

c. Sharp AL-1530CS Digital Copier 

```python
Image(filename='ques5.png')
```

<a id='6'></a>
**Question 6:**

In the Technology Product Category, which unprofitable state is surrounded by only profitable states?  

**Solution Steps:**

* Filter “Product Category” to Technology 
* Drag “State or Province” to Marks 
* Drag “Profit” to color
* Right click color, edit colors, stepped color: 2 steps, click OK

**Answer**: 

a. Colorado

```python
Image(filename='ques6.png')
```

<a id='7'></a>
**Question 7:**

If 2013 Sales numbers were expected to increase by 50% in the following year, what would be the total estimated sales for the Consumer Segment in 2014? 

**Solution Steps:**
* Create Calculated Field called Sales Increase: Sum([Sales]) * 1.5
* Filter “Customer Segment” to Consumer 
* 2nd Filter “Order Date” to 2013
* Drag Sales Increase to Row (will show up as AGG(Sales Increase))
* Filter Marks to Bar
* Click check box for show mark labels on Label 

**Answer:** 

b. $816,999

```python
Image(filename='ques7.png', width=500)
```

<a id='8'></a>
**Question 8:**

In which Region do all Product Categories fall beneath the overall average profit? 

**Solution Steps:**

* Drag "Region" to Columns
* Also drag "Product Category" to Columns
* Drag "Profit" to Rows
* Click Analytics Tab (top left, next to Data tab), drag Average Line to Bar Chart

**Answer:**

d. South

<a id='9'></a>
**Question 9:**

Which Product Sub-Category has a Shipping Cost to Sales ratio of above 3%?

**Solution Steps:**
* Create Calculated Field called Shipping Cost to Sales: SUM([Shipping Cost])/SUM([Sales])
* Drag "Product Sub-Category" to Columns
* Drag "Shipping Cost to Sales" to Rows, right click, choose range of values, 0.03 to 0.04, click OK 

**Answer:**
c. Paper

```python
Image(filename='ques9.png', width=400)
```

<a id='10'></a>
**Question 10:**

Find the customer with the lowest overall profit. What is his/her profit ratio?

**Solution Steps:**

* Create Calculator Field called Profit Ratio: SUM([Profit])/SUM([Sales])
* Right click Profit Ratio, Select Default Properties, Select Number Format, Select Percentage, Click OK 
* Drag "Customer ID" and "Customer Name" to Rows
* Drag "Profit" to Columns
* Drag "Profit Ratio" to Label 

**Answer:**

d. -771.39%

<a id='13'></a>
**Question 13:**  What percent of the total profit do the top 10 customers by Sales represent?

**Solution Steps:**

* Right click Customer ID, choose Create, Set, Name: Top 10 Customers by Sales, Select Top Tab, By field: Top 10 by Sales Sum, Click OK
* Drag "Profit" to Rows, click down arrow, go to Quick Table Calculation, click Percent of Total
* Drag the "Top 10 Customers by Sales" set to Color 

**Answer:**

b. 5.03%

```python
Image(filename='ques13.png', width=900)
```

<a id='14'></a>
**Question 14:** What was the Moving Average of Sales in June of 2012, including six months prior and six months after?

**Solution Steps:**
* Right click and drag "Order Date" to Columns, select MONTH(Order Date) with the calendar symbol beside it (a continuous variable) 
* Drag "Sales" to Rows, click down arrow, go to Quick Table Calculation, select Moving Average
* Click down arrow again on Sales, Edit Table Calculation, go to 2nd dropdown box, change previous values to 6 and next values to 6 

**Answer:**

c. 188,552

```python
Image(filename='ques14.png', width=900)
```

<a id='15'></a>
**Question 15:**  Create a histogram showing the number of Sales using Sales Bins of $1,000. Which bins have profit ratios (profit as a percentage of sales) of more than 25%? (Select all that apply)

**Solution Steps:**

* Create Calculator Field called Profit Ratio: SUM([Profit])/SUM([Sales])
* Right click on "Sales", go to Create, select Bins, Size of bins: 1,000, click OK
* Drag "Sales(bin)" to Columns
* Drag "Profit Ratio" to Rows
* Filter "AGG(Profit Ratio)" to at least 0.25

**Answer:**

7k, 8k, 11k, 18k
